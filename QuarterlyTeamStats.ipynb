```python
# Quarter 1 FG% for each game

quarter1_df = df[df["Quarter"] == 1]

for game_id, gameRow in games.iterrows():
    gameid = gameRow["Games"]
    # Filter rows for the current game
    game_data = quarter1_df[quarter1_df["URL"] == gameid]
    
    home_team_total_shots = 0
    home_team_made_shots = 0
    away_team_total_shots = 0
    away_team_made_shots = 0

    home_team_total_3_attempt = 0
    home_team_made_3_attempt = 0
    away_team_total_3_attempt = 0
    away_team_made_3_attempt = 0

    for _, row in game_data.iterrows():
        if pd.isnull(row['ShotType']):
            continue

        # Update home team FG%
        if "2-pt" in row['ShotType'] and pd.isnull(row['AwayPlay']):
            home_team_total_shots += 1
            if row['ShotOutcome'] == "make":
                home_team_made_shots += 1

        if "3-pt" in row['ShotType'] and pd.isnull(row['AwayPlay']):
            home_team_total_shots += 1
            home_team_total_3_attempt += 1
            if row['ShotOutcome'] == "make":
                home_team_made_shots += 1
                home_team_made_3_attempt += 1

        # Update away team FG%
        if "2-pt" in row['ShotType'] and pd.isnull(row['HomePlay']):
            away_team_total_shots += 1
            if row['ShotOutcome'] == "make":
                away_team_made_shots += 1
        if "3-pt" in row['ShotType'] and pd.isnull(row['HomePlay']):
            away_team_total_shots += 1
            away_team_total_3_attempt += 1
            if row['ShotOutcome'] == "make":
                away_team_made_shots += 1
                away_team_made_3_attempt += 1

    # Calculate FG%
    Q1_HomeTeam_FG_PER = (home_team_made_shots / home_team_total_shots if home_team_total_shots > 0 else 0)
    Q1_AwayTeam_FG_PER = (away_team_made_shots / away_team_total_shots if away_team_total_shots > 0 else 0)

    # Calculate 3P%
    Q1_HomeTeam_3P_PER = (home_team_made_3_attempt / home_team_total_3_attempt if home_team_total_3_attempt > 0 else 0)

    Q1_AwayTeam_3P_PER = (away_team_made_3_attempt / away_team_total_3_attempt if away_team_total_3_attempt > 0 else 0)


    # Add FG% to the games DataFrame
    games.loc[game_id, "Q1_HomeTeam_FG_Made"] = home_team_made_shots
    games.loc[game_id, "Q1_HomeTeam_FG_Total"] = home_team_total_shots
    games.loc[game_id, "Q1_HomeTeam_FG_%"] = Q1_HomeTeam_FG_PER

    games.loc[game_id, "Q1_AwayTeam_FG_Made"] = away_team_made_shots
    games.loc[game_id, "Q1_AwayTeam_FG_Total"] = away_team_total_shots
    games.loc[game_id, "Q1_AwayTeam_FG_%"] = Q1_AwayTeam_FG_PER


    # Add 3P% to the games DataFrame
    games.loc[game_id, "Q1_HomeTeam_3P_Made"] = home_team_made_3_attempt
    games.loc[game_id, "Q1_HomeTeam_3P_Total"] = home_team_total_3_attempt
    games.loc[game_id, "Q1_HomeTeam_3P_%"] = Q1_HomeTeam_3P_PER
    
    games.loc[game_id, "Q1_AwayTeam_3P_Made"] = away_team_made_3_attempt
    games.loc[game_id, "Q1_AwayTeam_3P_Total"] = away_team_total_3_attempt
    games.loc[game_id, "Q1_AwayTeam_3P_%"] = Q1_AwayTeam_3P_PER



games.head()
```

```python
# Calculate FT% For Each Game
quarter1_df = df[df["Quarter"] == 1]

for game_id, gameRow in games.iterrows():
    home_team_FT_total_attempts = 0
    home_team_FT_made_attempts = 0

    away_team_FT_total_attempts = 0
    away_team_FT_made_attempts = 0

    gameid = gameRow["Games"]

    game_data = quarter1_df[quarter1_df["URL"] == gameid]

    for _, row in game_data.iterrows():
        if pd.isnull(row['FreeThrowNum']):
            continue

        # FT is shot
        else:
            # Home Team Shot It
            if pd.isnull(row['AwayPlay']):
                home_team_FT_total_attempts += 1
                if row['FreeThrowOutcome'] == 'make':
                    home_team_FT_made_attempts += 1

            # Away Team Shot It
            if pd.isnull(row['HomePlay']):
                away_team_FT_total_attempts += 1
                if row['FreeThrowOutcome'] == 'make':
                    away_team_FT_made_attempts += 1

    # Calculate FT% after processing all rows for the game
    Q1_HomeTeam_FT_PER = (home_team_FT_made_attempts / home_team_FT_total_attempts if home_team_FT_total_attempts > 0 else 0)
    Q1_AwayTeam_FT_PER = (away_team_FT_made_attempts / away_team_FT_total_attempts if away_team_FT_total_attempts > 0 else 0)

    # Update the games DataFrame


    games.loc[game_id, "Q1_HomeTeam_FT_Made"] = home_team_FT_made_attempts
    games.loc[game_id, "Q1_HomeTeam_FT_Total"] = home_team_FT_total_attempts
    games.loc[game_id, "Q1_HomeTeam_FT_%"] = Q1_HomeTeam_FT_PER

    games.loc[game_id, "Q1_AwayTeam_FT_Made"] = away_team_FT_made_attempts
    games.loc[game_id, "Q1_AwayTeam_FT_Total"] = away_team_FT_total_attempts
    games.loc[game_id, "Q1_AwayTeam_FT_%"] = Q1_AwayTeam_FT_PER


games.head()
```

```python
#Q1 Offensive Rebound and Defensive Rebound
quarter1_df = df[df["Quarter"] == 1]

for game_id, gameRow in games.iterrows():

  home_team_OR_total = 0
  home_team_DR_total = 0
  home_team_total_rebounds = 0

  away_team_OR_total = 0
  away_team_DR_total = 0
  away_team_total_rebounds = 0

  gameid = gameRow["Games"]

  game_data = quarter1_df[quarter1_df["URL"] == gameid]

  for _, row in game_data.iterrows():
    
      if row['ReboundType'] == "offensive":
        if pd.isnull(row['AwayPlay']):
          if "Team" not in row['HomePlay']:
            home_team_OR_total += 1
        elif pd.isnull(row['HomePlay']):
          if "Team" not in row['AwayPlay']:
            away_team_OR_total += 1

      elif row['ReboundType'] == "defensive":
        if pd.isnull(row['AwayPlay']):
          if "Team" not in row['HomePlay']:
            home_team_DR_total += 1
        elif pd.isnull(row['HomePlay']):
          if "Team" not in row['AwayPlay']:
            away_team_DR_total += 1

        

  games.loc[game_id, "Q1_HomeTeam_Total_Rebounds"] = home_team_OR_total + home_team_DR_total 
  games.loc[game_id, "Q1_HomeTeam_OR"] = home_team_OR_total
  games.loc[game_id, "Q1_HomeTeam_DR"] = home_team_DR_total

  
  games.loc[game_id, "Q1_AwayTeam_Total_Rebounds"] = away_team_OR_total + away_team_DR_total 
  games.loc[game_id, "Q1_AwayTeam_OR"] = away_team_OR_total
  games.loc[game_id, "Q1_AwayTeam_DR"] = away_team_DR_total

games.head()
```

```python

#Q1 Assists
quarter1_df = df[df["Quarter"] == 1]

for game_id, gameRow in games.iterrows():


  home_team_total_assists = 0
  
  away_team_total_assists = 0

  gameid = gameRow["Games"]

  game_data = quarter1_df[quarter1_df["URL"] == gameid]

  for _, row in game_data.iterrows():
    
      if pd.notnull(row["Assister"]) and pd.notnull(row["HomePlay"]):
        home_team_total_assists += 1
      if pd.notnull(row["Assister"]) and pd.notnull(row["AwayPlay"]):
        away_team_total_assists += 1

      

        

  games.loc[game_id, "Q1_HomeTeam_Total_Assists"] =  home_team_total_assists
 
  games.loc[game_id, "Q1_AwayTeam_Total_Assists"] = away_team_total_assists
  

games.head()
```

```python
#Q2 Offensive Rebound and Defensive Rebound / Assits
quarter2_df = df[df["Quarter"] == 2]

for game_id, gameRow in games.iterrows():

  # Game ID
  gameid = gameRow["Games"]

  #Q2 Offensive Rebound and Defensive Rebound
  home_team_OR_total = 0
  home_team_DR_total = 0
  home_team_total_rebounds = 0
  away_team_OR_total = 0
  away_team_DR_total = 0
  away_team_total_rebounds = 0

  # Q2 Assists
  home_team_total_assists = 0
  away_team_total_assists = 0

  # Q2 Steals
  home_team_total_steals = 0
  away_team_total_steals = 0

  # Q2 Blocks
  home_team_total_blocks = 0  
  away_team_total_blocks = 0

  # Q2 Turnover
  home_team_total_turnovers = 0
  away_team_total_turnovers = 0

  # Q2 Score
  home_team_q2_score = 0
  away_team_q2_score = 0

 

  # Filter rows for the current game
  game_data = quarter2_df[quarter2_df["URL"] == gameid]

  for _, row in game_data.iterrows():
    
      # Q2 Offensive Rebound and Defensive Rebound
      if row['ReboundType'] == "offensive":
        if pd.isnull(row['AwayPlay']):
          if "Team" not in row['HomePlay']:
            home_team_OR_total += 1
        elif pd.isnull(row['HomePlay']):
          if "Team" not in row['AwayPlay']:
            away_team_OR_total += 1

      elif row['ReboundType'] == "defensive":
        if pd.isnull(row['AwayPlay']):
          if "Team" not in row['HomePlay']:
            home_team_DR_total += 1
        elif pd.isnull(row['HomePlay']):
          if "Team" not in row['AwayPlay']:
            away_team_DR_total += 1

      # Q2 Assists
      if pd.notnull(row["Assister"]) and pd.notnull(row["HomePlay"]):
        home_team_total_assists += 1
      if pd.notnull(row["Assister"]) and pd.notnull(row["AwayPlay"]):
        away_team_total_assists += 1

      # Q2 Steals
      if pd.notnull(row["TurnoverCause"]) and pd.notnull(row["HomePlay"]):
        home_team_total_steals += 1
      if pd.notnull(row["TurnoverCause"]) and pd.notnull(row["AwayPlay"]):
        away_team_total_steals += 1

      # Q2 Blocks
      if pd.notnull(row["Blocker"]) and pd.notnull(row["HomePlay"]):
        home_team_total_blocks += 1
      if pd.notnull(row["Blocker"]) and pd.notnull(row["AwayPlay"]):
        away_team_total_blocks += 1

      # Q2 Turnovers
      if pd.notnull(row["TurnoverPlayer"]):
        if pd.notnull(row["HomePlay"]):
          home_team_total_turnovers += 1
        elif pd.notnull(row["AwayPlay"]):
          away_team_total_turnovers += 1

      # Q2 Score
      if row['AwayPlay'] == "End of 2nd quarter":
        home_team_q2_score = row['HomeScore']
        away_team_q2_score = row['AwayScore']

  # Q2 Offensive Rebound and Defensive Rebound
  games.loc[game_id, "Q2_HomeTeam_Total_Rebounds"] = home_team_OR_total + home_team_DR_total 
  games.loc[game_id, "Q2_HomeTeam_OR"] = home_team_OR_total
  games.loc[game_id, "Q2_HomeTeam_DR"] = home_team_DR_total

  
  games.loc[game_id, "Q2_AwayTeam_Total_Rebounds"] = away_team_OR_total + away_team_DR_total 
  games.loc[game_id, "Q2_AwayTeam_OR"] = away_team_OR_total
  games.loc[game_id, "Q2_AwayTeam_DR"] = away_team_DR_total
  
  # Q2 Assists
  games.loc[game_id, "Q2_HomeTeam_Total_Assists"] = home_team_total_assists
  games.loc[game_id, "Q2_AwayTeam_Total_Assists"] = away_team_total_assists

  # Q2 Steals
  games.loc[game_id, "Q2_HomeTeam_Total_Steals"] = home_team_total_steals
  games.loc[game_id, "Q2_AwayTeam_Total_Steals"] = away_team_total_steals
  
  # Q2 Blocks
  games.loc[game_id, "Q2_HomeTeam_Total_Blocks"] = home_team_total_blocks
  games.loc[game_id, "Q2_AwayTeam_Total_Blocks"] = away_team_total_blocks

  # Q2 Turnovers
  games.loc[game_id, "Q2_HomeTeam_Total_Turnovers"] = home_team_total_turnovers
  games.loc[game_id, "Q2_AwayTeam_Total_Turnovers"] = away_team_total_turnovers
  
  # Q2 Score
  Q1_HomeTeam_Score_Temp = games.loc[game_id, "Q1_HomeTeam_Score"] 
  Q1_AwayTeam_Score_Temp = games.loc[game_id, "Q1_AwayTeam_Score"] 
  games.loc[game_id, "Q2_HomeTeam_Score"] = home_team_q2_score - Q1_HomeTeam_Score_Temp
  games.loc[game_id, "Q2_HomeTeam_Point_Dif"] = (home_team_q2_score - Q1_HomeTeam_Score_Temp) - (away_team_q2_score - Q1_AwayTeam_Score_Temp)

  games.loc[game_id, "Q2_AwayTeam_Score"] = away_team_q2_score - Q1_AwayTeam_Score_Temp
  games.loc[game_id, "Q2_AwayTeam_Point_Dif"] = (away_team_q2_score - Q1_AwayTeam_Score_Temp) - (home_team_q2_score - Q1_HomeTeam_Score_Temp) 
# Print Games Data Frame 
games.head()
```

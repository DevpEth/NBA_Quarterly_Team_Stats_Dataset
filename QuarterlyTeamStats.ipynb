```python
# Quarter 1 FG% for each game

quarter1_df = df[df["Quarter"] == 1]

for game_id, gameRow in games.iterrows():
    gameid = gameRow["Games"]
    # Filter rows for the current game
    game_data = quarter1_df[quarter1_df["URL"] == gameid]
    
    home_team_total_shots = 0
    home_team_made_shots = 0
    away_team_total_shots = 0
    away_team_made_shots = 0

    home_team_total_3_attempt = 0
    home_team_made_3_attempt = 0
    away_team_total_3_attempt = 0
    away_team_made_3_attempt = 0

    for _, row in game_data.iterrows():
        if pd.isnull(row['ShotType']):
            continue

        # Update home team FG%
        if "2-pt" in row['ShotType'] and pd.isnull(row['AwayPlay']):
            home_team_total_shots += 1
            if row['ShotOutcome'] == "make":
                home_team_made_shots += 1

        if "3-pt" in row['ShotType'] and pd.isnull(row['AwayPlay']):
            home_team_total_shots += 1
            home_team_total_3_attempt += 1
            if row['ShotOutcome'] == "make":
                home_team_made_shots += 1
                home_team_made_3_attempt += 1

        # Update away team FG%
        if "2-pt" in row['ShotType'] and pd.isnull(row['HomePlay']):
            away_team_total_shots += 1
            if row['ShotOutcome'] == "make":
                away_team_made_shots += 1
        if "3-pt" in row['ShotType'] and pd.isnull(row['HomePlay']):
            away_team_total_shots += 1
            away_team_total_3_attempt += 1
            if row['ShotOutcome'] == "make":
                away_team_made_shots += 1
                away_team_made_3_attempt += 1

    # Calculate FG%
    Q1_HomeTeam_FG_PER = (home_team_made_shots / home_team_total_shots if home_team_total_shots > 0 else 0)
    Q1_AwayTeam_FG_PER = (away_team_made_shots / away_team_total_shots if away_team_total_shots > 0 else 0)

    # Calculate 3P%
    Q1_HomeTeam_3P_PER = (home_team_made_3_attempt / home_team_total_3_attempt if home_team_total_3_attempt > 0 else 0)

    Q1_AwayTeam_3P_PER = (away_team_made_3_attempt / away_team_total_3_attempt if away_team_total_3_attempt > 0 else 0)


    # Add FG% to the games DataFrame
    games.loc[game_id, "Q1_HomeTeam_FG_Made"] = home_team_made_shots
    games.loc[game_id, "Q1_HomeTeam_FG_Total"] = home_team_total_shots
    games.loc[game_id, "Q1_HomeTeam_FG_%"] = Q1_HomeTeam_FG_PER

    games.loc[game_id, "Q1_AwayTeam_FG_Made"] = away_team_made_shots
    games.loc[game_id, "Q1_AwayTeam_FG_Total"] = away_team_total_shots
    games.loc[game_id, "Q1_AwayTeam_FG_%"] = Q1_AwayTeam_FG_PER


    # Add 3P% to the games DataFrame
    games.loc[game_id, "Q1_HomeTeam_3P_Made"] = home_team_made_3_attempt
    games.loc[game_id, "Q1_HomeTeam_3P_Total"] = home_team_total_3_attempt
    games.loc[game_id, "Q1_HomeTeam_3P_%"] = Q1_HomeTeam_3P_PER
    
    games.loc[game_id, "Q1_AwayTeam_3P_Made"] = away_team_made_3_attempt
    games.loc[game_id, "Q1_AwayTeam_3P_Total"] = away_team_total_3_attempt
    games.loc[game_id, "Q1_AwayTeam_3P_%"] = Q1_AwayTeam_3P_PER



games.head()
```
